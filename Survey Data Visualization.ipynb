# Survey Data Visualization Project

Loading Data

```python
import pandas as pd

df = pd.read_csv("survey_data_project8.csv")
print(df.head())
print(df.info())
```

# Creating Visuals

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data=df, x='Gender')
plt.title("Gender Distribution")
plt.show()
```

```python
df['Preferred Feature'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Preferred Feature Distribution")
plt.ylabel('')
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import plotly.express as px

fig = px.histogram(df, x="Satisfaction", color="Gender", barmode="group")
fig.show()
```

Most users prefer Feature A (40%)

Males spend more time than females on average
